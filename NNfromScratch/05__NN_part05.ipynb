```python
import numpy as np

# Initial parameters
initial_weights = np.array([-3.0, -1.0, 2.0])  # Weight values for inputs x0, x1, x2
initial_bias = 1.0  # Bias term
input_values = np.array([1.0, -2.0, 3.0])  # Input values x0, x1, x2
target_value = 0.0  # Desired output value
learning_rate = 0.001  # Step size for weight updates

# Activation function: Returns max(0, x) to introduce non-linearity
def relu(activation_input):
    return np.maximum(0, activation_input)

# Derivative of ReLU: 1 if input > 0, 0 otherwise
def relu_derivative(activation_input):
    return np.where(activation_input > 0, 1.0, 0.0)

# Training loop for 100 iterations
for iteration in range(100):
    # Forward pass: Compute the prediction
    # Calculate the weighted sum of inputs plus bias
    weighted_sum = np.dot(initial_weights, input_values) + initial_bias
    # Apply ReLU activation to get the output
    predicted_output = relu(weighted_sum)
    # Compute the loss (mean squared error)
    loss = (predicted_output - target_value) ** 2

    # Backward pass: Compute gradients
    # Gradient of loss with respect to output: 2 * (output - target)
    gradient_loss_output = 2 * (predicted_output - target_value)
    # Gradient of output with respect to weighted sum (ReLU derivative)
    gradient_output_weighted_sum = relu_derivative(weighted_sum)
    # Gradient of weighted sum with respect to weights (equals inputs)
    gradient_weighted_sum_weights = input_values
    # Gradient of weighted sum with respect to bias (always 1)
    gradient_weighted_sum_bias = 1.0

    # Combine gradients using chain rule
    # Gradient of loss with respect to weighted sum
    gradient_loss_weighted_sum = gradient_loss_output * gradient_output_weighted_sum
    # Gradient of loss with respect to weights
    gradient_loss_weights = gradient_loss_weighted_sum * gradient_weighted_sum_weights
    # Gradient of loss with respect to bias
    gradient_loss_bias = gradient_loss_weighted_sum * gradient_weighted_sum_bias

    # Update weights and bias using gradient descent
    initial_weights -= learning_rate * gradient_loss_weights
    initial_bias -= learning_rate * gradient_loss_bias

    # Print loss for each iteration to track progress
    print(f"Iteration {iteration + 1}, Loss: {loss}")

# Print final trained parameters
print(f"Final weights: {initial_weights}")
print(f"Final bias: {initial_bias}")
```
